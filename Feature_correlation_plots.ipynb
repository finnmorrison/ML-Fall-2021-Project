```python
# Pairwise covariance of columns
company_stock_sentiment.cov()
```

```python
# Pairwise correlation of columns (using standard correlation coefficient)
company_values_corr = company_stock_sentiment.corr(method='pearson')
company_stock_sentiment.corr(method='pearson')
```

```python
# Correlation heatmap for features

# http://seaborn.pydata.org/examples/many_pairwise_correlations.html

# Generate a mask for the upper triangle
mask = np.zeros_like(company_values_corr, dtype=np.bool_)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(8, 5))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
heatmap = sns.heatmap(company_values_corr, mask=mask, cmap=cmap, vmax=.3,
            square=True, linewidths=.5, cbar_kws={"label": 'correlation',"shrink": .5}, ax=ax, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```
